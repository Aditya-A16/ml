```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
#Generate a dummy dataset.
X = np.random.randint(1,10,60).reshape(20,3) 
# mean Centering the data  
X_meaned = X - np.mean(X , axis = 0)
```

```python
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")
ax.scatter3D(X[:, :1], X[:, 1:2], X[:, 2:3])
```

```python
cov_mat = np.cov(X_meaned.T)
```

```python
eigen_values , eigen_vectors = np.linalg.eigh(cov_mat)
```

```python
eigen_vectors = eigen_vectors.T
```

```python
sorted_index = np.argsort(eigen_values)[::-1]


sorted_eigenvalue = eigen_values[sorted_index]
sorted_eigenvectors = eigen_vectors[sorted_index]
```

```python
n_components = 2
components = sorted_eigenvectors[0:n_components]  
```

```python
#Transform the data 
X_reduced = np.dot(X_meaned, components.T)
```

```python
X_reduced
plt.scatter(X_reduced[:, :1],X_reduced[:, 1:])
```

```python
#decompression
x_approx = np.dot(  X_reduced, components)
```
